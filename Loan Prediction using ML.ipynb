```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df=pd.read_excel('loan status.csv.xlsx')
df
```

```python
df.columns
```

```python
X=df[['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
```

```python
y=df.Loan_Status
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_test.shape
```

### Preprocessing train set

```python
X_train.isnull().sum()
```

```python
# filling all nan values of categorical variables

gmode=X_train['Gender'].mode().iloc[0]
dmode=X_train['Dependents'].mode().iloc[0]
mmode=X_train['Married'].mode().iloc[0]
smode=X_train['Self_Employed'].mode().iloc[0]
cmode=X_train['Credit_History'].mode().iloc[0]
lamode=X_train['Loan_Amount_Term'].mode().iloc[0]


X_train['Gender'].fillna(gmode,inplace=True)
X_train['Dependents'].fillna(dmode,inplace=True)
X_train['Married'].fillna(mmode,inplace=True)
X_train['Self_Employed'].fillna(smode,inplace=True)
X_train['Credit_History'].fillna(cmode,inplace=True)
X_train['Loan_Amount_Term'].fillna(lamode,inplace=True)
```

```python
# filling nan values of numerical variables

X_train['LoanAmount'].fillna(X_train['LoanAmount'].median(),inplace=True)
```

```python
X_train.isnull().sum()
```

```python
# Transforming the values of dependents from 3+ to 3

X_train.loc[X_train['Dependents']=='3+','Dependents']=3
X_train['Dependents'].value_counts()
```
